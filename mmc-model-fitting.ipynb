```python
#!pip install metric-learn
import pandas as pd
import numpy as np
from metric_learn import MMC
from sklearn.preprocessing import MinMaxScaler
import pickle
import time

# Record start time
start_time = time.time()

# Assuming your CSV file is named 'combined_results.csv', you would replace this with the actual filename
file_path = '/kaggle/input/training-data-set/training.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Extract columns for scaling
columns_to_scale = ['X1', 'Y1', 'Z1', 'T1', 'X2', 'Y2', 'Z2', 'T2']
data_to_scale = df[columns_to_scale]

# Scale the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data_to_scale)

# Reshape the scaled data to the desired format
scaled_data_reshaped = scaled_data.reshape(len(df), 2, 4)

# Now you can use the reshaped scaled data directly for fitting the MMC model
mmc = MMC()
mmc.fit(scaled_data_reshaped, df['Label'])


with open('mmc_model_scaled.pkl', 'wb') as f:
    pickle.dump(mmc, f)

# Calculate execution time
execution_time = time.time() - start_time
print("Execution time:", execution_time, "seconds")
```
